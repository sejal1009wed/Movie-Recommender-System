```python
import pandas as pd
```

```python
df = pd.read_csv('imdb_top_1000.csv')
df.head(5)
```

```python
df = df[['Poster_Link','Series_Title','Genre','IMDB_Rating','Overview','Director','Star1','Star2','Star3','Star4']]
df.head(2)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop(['IMDB_Rating'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df['Director'] = df['Director'].apply(lambda x:x.split())
df['Genre'] = df['Genre'].apply(lambda x:x.split())
df['Stars'] = df['Star1'].apply(lambda x:x.split()) + df['Star2'].apply(lambda x:x.split()) + df['Star3'].apply(lambda x:x.split())+ df['Star4'].apply(lambda x:x.split())
```

```python
df.head(3)
```

```python
df.drop(['Star1','Star2','Star3','Star4'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['tags'] = df['Genre']+ df['Overview']+ df['Director']+ df['Stars']
```

```python
df.head(2)
```

```python
df['tags'] = df['tags'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head(2)
```

```python
new_df = df[['Poster_Link','Series_Title','tags']]
```

```python
new_df
```

```python
#Converting the list type of tags to string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(2)
```

```python
new_df['tags'][1]
```

```python
#Lowercasing the tags column as its a better practice
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
#Applying vectorization 
!pip install scikit-learn
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv =  CountVectorizer(max_features=1000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
#actions or action are same but classified as two different features this is not good
# for optimisation we have to remove this with stemming
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#Fetching index of movie
#Example
new_df[new_df['Series_Title']=='Giant'].index[0]

#Index of moviws ko hold kerke rakhna  h use enumarate 
#sorted(list(enumerate(similarity[0])),reverse=True,key= lambda x:x[1])
```

```python
def recommend(movie):
    movie_index = new_df[new_df['Series_Title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].Series_Title)
```

```python
recommend('Giant')
```
